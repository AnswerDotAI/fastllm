```
import os, asyncio
from fastllm import OpenAIClient, AnthropicClient, GeminiClient, Msg, Part
```

```
OPENAI_MODEL = "gpt-5-mini"
ANTHROPIC_MODEL = "claude-sonnet-4-5"
GEMINI_MODEL = "gemini-2.5-flash"
```

```
openai = OpenAIClient(api_key=os.getenv("OPENAI_API_KEY",""))
anth = AnthropicClient(api_key=os.getenv("ANTHROPIC_API_KEY",""))
gem = GeminiClient(api_key=os.getenv("GEMINI_API_KEY",""))
```

### OpenAI
